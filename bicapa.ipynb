## Membrana curva (esfera / cilindro / silla)
Proyecta la bicapa sobre una superficie curva y exporta PDB + MRC + YAML para PolNet.

```python
from curved_geometry import (
    spherical_patch, cylindrical_patch, saddle_patch,
    curvature_stability_scan, plot_curvature_analysis
)
from export_curved import export_curved_all

b = BicapaCryoET(size_nm=SIZE_NM, seed=SEED).build()

print('=== Analisis de estabilidad de curvatura ===')
scan = curvature_stability_scan(b, [25, 50, 100, 200])
for d in scan:
    print('  R=%-5d %-10s E=%.3f kBT/lip [%s]' % (
        d['radius_nm'], d['geometry'],
        d['E_per_lipid_kBT'], d['stable']))
```

```python
plot_curvature_analysis(b)
```

```python
# Cilindro R=25nm (reticulo endoplasmatico)
r_cyl = export_curved_all(b, cylindrical_patch(25.0, SIZE_NM))
print('Cilindro exportado:')
for k, v in r_cyl.items():
    if isinstance(v, str): print('  %s: %s' % (k, v))
```

```python
# Esfera R=50nm (vesicula de secrecion)
r_sph = export_curved_all(b, spherical_patch(50.0, SIZE_NM))
print('Esfera exportada:')
for k, v in r_sph.items():
    if isinstance(v, str): print('  %s: %s' % (k, v))
```

```python
# Silla de montar R=50nm (cuello de fusion)
r_sad = export_curved_all(b, saddle_patch(50.0, SIZE_NM))
print('Silla exportada:')
for k, v in r_sad.items():
    if isinstance(v, str): print('  %s: %s' % (k, v))

print()
print('Para usar con PolNet:')
print('  polnet --config', r_cyl['yaml'])
```

## Inspeccionar propiedades de la bicapa

```python
b = BicapaCryoET(size_nm=SIZE_NM, seed=SEED).build()

print('Lípidos totales:  ', len(b.outer_leaflet) + len(b.inner_leaflet))
print('Monocapa externa: ', len(b.outer_leaflet))
print('Monocapa interna: ', len(b.inner_leaflet))
print('Balsas externas:  ', len(b.rafts_outer))
print('Balsas internas:  ', len(b.rafts_inner))
print('Clusters PIPs:    ', len(b.pip_clusters))
print()
print('kc (kBT·nm²):     ', round(b.bending_modulus, 1))
print('sigma (kBT/nm²):  ', round(b.surface_tension, 4))
print()
print('Grosor total (Å): ', round(b.geometry.total_thick, 1))
print('Grosor hidro (Å): ', round(b.geometry.hydro_thick, 1))
print()
print('Composición externa:')
for k, v in sorted(b.comp_outer.items(), key=lambda x: -x[1]):
    print('  %-8s %.1f%%' % (k, v * 100))
print('Composición interna:')
for k, v in sorted(b.comp_inner.items(), key=lambda x: -x[1]):
    print('  %-8s %.1f%%' % (k, v * 100))
```
