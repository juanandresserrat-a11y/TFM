## Inspeccionar propiedades de la bicapa

```python
b = BicapaCryoET(size_nm=SIZE_NM, seed=SEED).build()

print('Lípidos totales:  ', len(b.outer_leaflet) + len(b.inner_leaflet))
print('Monocapa externa: ', len(b.outer_leaflet))
print('Monocapa interna: ', len(b.inner_leaflet))
print('Balsas externas:  ', len(b.rafts_outer))
print('Balsas internas:  ', len(b.rafts_inner))
print('Clusters PIPs:    ', len(b.pip_clusters))
print()
print('kc (kBT·nm²):     ', round(b.bending_modulus, 1))
print('sigma (kBT/nm²):  ', round(b.surface_tension, 4))
print()
print('Grosor total (Å): ', round(b.geometry.total_thick, 1))
print('Grosor hidro (Å): ', round(b.geometry.hydro_thick, 1))
print()
print('Composición externa:')
for k, v in sorted(b.comp_outer.items(), key=lambda x: -x[1]):
    print('  %-8s %.1f%%' % (k, v * 100))
print('Composición interna:')
for k, v in sorted(b.comp_inner.items(), key=lambda x: -x[1]):
    print('  %-8s %.1f%%' % (k, v * 100))
```
